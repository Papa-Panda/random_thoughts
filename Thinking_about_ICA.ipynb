```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.decomposition import FastICA
import pandas as pd
import seaborn as sns

# Load the iris dataset
data = load_iris()
X = data.data
y = data.target
feature_names = data.feature_names

# Apply ICA
ica = FastICA(n_components=3, random_state=0)
X_ica = ica.fit_transform(X)

# Create a DataFrame for visualization
df = pd.DataFrame(X_ica, columns=[f"ICA{i+1}" for i in range(X_ica.shape[1])])
df['species'] = y

# Plot the results
plt.figure(figsize=(10, 7))
sns.scatterplot(data=df, x='ICA1', y='ICA2', hue='species', palette='viridis')
plt.title('ICA applied to Iris dataset')
plt.show()

# Plot the original features for comparison
df_original = pd.DataFrame(X, columns=feature_names)
df_original['species'] = y

plt.figure(figsize=(10, 7))
sns.pairplot(df_original, hue='species', palette='viridis')
plt.suptitle('Original Iris dataset features', y=1.02)
plt.show()
```
